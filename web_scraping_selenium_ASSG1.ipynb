# Q2: Write a python program to scrape data for “Data Scientist” Job position in 
“Bangalore” location. You have to scrape the job-title, job-location,
company_name, full job-description. You have to scrape first 10 jobs data.

```python
driver=webdriver.Chrome("chromedriver.exe")#opening the chrome driver
```

```python
driver.get("https://www.naukri.com/")
```

```python
search_job=driver.find_element_by_id("qsb-keyword-sugg")
search_job.send_keys("Data Scientist")
search_loc=driver.find_element_by_id("qsb-location-sugg")
search_loc.send_keys("Bangalore")
```

```python
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
job_desc=[]
job_url=[]
```

```python
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):

    job_url.append(i.get_attribute('href'))
```

```python

for i in job_url[:10]:

    driver.get(i)

    try:

        title_tag = driver.find_element_by_xpath("//h1[@class='jd-header-title']")

        job_title.append(title_tag.text)

    except:

        job_title.append('-')

    try:

        loc_tag = driver.find_element_by_xpath("//span[@class='location ']")

        job_location.append(loc_tag.text)

    except:

        job_location.append('-')

    try:

        cp_tag = driver.find_element_by_xpath("//a[@class='pad-rt-8']")

        company_name.append(cp_tag.text)

    except:

        company_name.append('-')
        
    try:

        dc_tag = driver.find_element_by_xpath("//section[@class='job-desc']")

        job_desc.append(dc_tag.text)

    except:

        job_desc.append('-')
        
```

```python
job_title
```

```python
desc=[]
for i in job_desc:
    desc.append(i.replace('\n',''))
desc
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['Titles']=job_title
jobs['Location']=job_location
jobs['Company']=company_name
jobs['Job description']=desc
jobs
```

```python
driver.close()
```
