1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome("chromedriver.exe")#opening the chrome driver
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[]
name=[]
artist=[]
uploaddate=[]
views=[]
```

```python
for i in driver.find_elements_by_xpath("//div[@class='mw-body-content mw-content-ltr']/div/table[3]/tbody/tr/td[1]"):
    rank.append(i.text)
    
for j in driver.find_elements_by_xpath("//div[@class='mw-body-content mw-content-ltr']/div/table[3]/tbody/tr/td[2]"):
    name.append(j.text)
    
for k in driver.find_elements_by_xpath("//div[@class='mw-body-content mw-content-ltr']/div/table[3]/tbody/tr/td[3]"):
    artist.append(k.text)
    
for l in driver.find_elements_by_xpath("//div[@class='mw-body-content mw-content-ltr']/div/table[3]/tbody/tr/td[5]"):
    uploaddate.append(l.text)
    
for m in driver.find_elements_by_xpath("//div[@class='mw-body-content mw-content-ltr']/div/table[3]/tbody/tr/td[4]"):
    views.append(m.text)
```

```python
youtube=pd.DataFrame({'Rank':rank,
                     'Video Name':name,
                     'Artist':artist,
                     'Upload Date':uploaddate,
                     'Total Views(in billions)':views})
youtube
```

```python
driver.close()
```

2.Scrape the details team India’s international fixtures from bcci.tv.

You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.bcci.tv/")
```

```python
international=driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]")
international.click()
fixture=driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a")
fixture.click()
```

```python
match=[]
series=[]
place=[]
date=[]
time=[]
l1=[]
l2=[]
l3=[]
```

```python
for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__format']"):
    match.append(i.text)

for j in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
    series.append(j.text)
    
for k in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
    place.append(k.text)
    
for l in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']/span"):
    l1.append(l.text)
for m in driver.find_elements_by_xpath("//div[@class='fixture__full-date']/span"):
    l2.append(m.text)
for n in driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[1]"):
    l3.append(n.text)
date=[o + p + q for o,p,q in zip(l1,l2,l3)]
```

```python
page_url=[]

for r in driver.find_elements_by_xpath('//div[@class="js-list"]/a'):



    page_url.append(r.get_attribute('href'))
```

```python
time=[]
for i in page_url:
    
    driver.get(i)
    
    try:
        s=driver.find_element_by_xpath("//span[@class='mc-header-scorebox__ist-time']")

        time.append(s.text)
    except:
        time.append('-')
```

```python
fixture=pd.DataFrame({'Match Title':match,
                     'Series':series,
                     'Place':place,
                     'Date':date,
                     'Time':time})
fixture
```

```python
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
name=[]
year=[]
genre=[]
run=[]
ratings=[]
votes=[]

for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
    name.append(i.text)

for j in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]"):
    year.append(j.text)
    
for k in driver.find_elements_by_xpath("//span[@class='genre']"):
    genre.append(k.text)
    
for l in driver.find_elements_by_xpath("//span[@class='runtime']"):
    run.append(l.text)

for m in driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']/div/span[2]"):
    ratings.append(m.text)

for n in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]"):
    votes.append(n.text)
    
imdb=pd.DataFrame({'Series name':name,
                  'Year span':year,
                  'Genre':genre,
                  "Runtime":run,
                  'Ratings':ratings,
                  'Votes':votes})
imdb
```

10. Details of Datasets from UCI machine learning repositories.
 Url = https://archive.ics.uci.edu/
 You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code

```python
driver.get(" https://archive.ics.uci.edu/")
```

```python
view_btn=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
view_btn.click()
```

```python
name=[]
data_type=[]
task=[]
attribute=[]
instance=[]
no_attri=[]
year=[]


for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]"):
    name.append(i.text)

for j in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]"):
    data_type.append(j.text)
    
for k in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]"):
    task.append(k.text)
    
for l in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]"):
    attribute.append(l.text)

for m in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]"):
    instance.append(m.text)

for n in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]"):
    no_attri.append(n.text)

for o in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]"):
    year.append(o.text)
    
uci=pd.DataFrame({'Dataset name':name,
                 'Data Type':data_type,
                 'Default Task':task,
                 'attribute':attribute,
                 'No. of Instance':instance,
                 "no. of attribute":no_attri,
                 'Year':year})
uci
```
