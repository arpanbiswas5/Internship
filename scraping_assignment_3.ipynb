vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of

```python
urls=[]
brand=[]
name=[]
rating=[]
no_rat=[]
price=[]
ret=[]
exp_deli=[]
availability=[]
other=[]
product_url=[]
page_url=[]
```

```python

for i in driver.find_elements_by_xpath('//li[@class="a-selected" or @class="a-normal"]/a'):



    page_url.append(i.get_attribute('href'))
```

```python
for i in page_url:
    
    driver.get(i)
    
    for i in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):

        product_url.append(i.get_attribute('href'))
```

```python
print(len(product_url))
```

```python
for i in product_url:

    driver.get(i)
    
    
    try:

        name_tag = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")

        name.append(name_tag.text)

    except:

        name.append('-')
        


    try:

        rating_tag = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")

        rating.append(rating_tag.text)

    except:

        rating.append('-')
    
    try:

        norat_tag = driver.find_element_by_xpath("//div[@class='a-row a-spacing-medium averageStarRatingNumerical']/span[1]")

        no_rat.append(norat_tag.text)

    except:

        no_rat.append('-')
        
    try:

        price_tag = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockDealPriceString']")

        price.append(price_tag.text)

    except:

        price.append('-')

    try:

        ret_tag = driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']/span/div[2]/a")

        ret.append(ret_tag.text)

    except:

        ret.append('-')
        
        
    try:

        del_tag = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b")

        exp_deli.append(del_tag.text)

    except:

        exp_deli.append('-')

    try:

        ava_tag = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-success']")

        availability.append(ava_tag.text)

    except:

        availability.append('-')
        
    try:

        other_tag = driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")

        other.append(other_tag.text)

    except:

        other.append('-')
        

    
    
    
```

```python
guitar=pd.DataFrame({})
guitar['product_name']=name
guitar['rating']=rating
guitar['no. of Ratings']=no_rat
guitar['price'] = price
guitar['Return/exchabge'] = ret
guitar['expected delivery'] = exp_deli
guitar['availability'] = availability
guitar['Other Details'] = other
guitar['product url']= product_url
guitar
```

```python
guitar.to_csv("guitarinfo.csv")
```
